**Objectives**
- Segment customers based on their *browsing time* and *purchase amount*.
- Apply **DBSCAN** clustering to identify natural customer groups without predefining the number of clusters.
- Tune DBSCAN hyperparameters (`eps`, `min_samples`) and visualize the clustering outcome.
- Interpret resulting clusters and detect outliers (noise points).

**Tools Used**
- Python Libraries: pandas, matplotlib, sklearn
- Methods:
  - DBSCAN clustering (sklearn.cluster.DBSCAN)
  - Scatter plot visualization for clusters
  - Manual tuning of `eps` and `min_samples`

#1. üì• Data Preparation

Data contains two characteristics:  
- `browsing_time` - the number of minutes spent browsing the store's webpage;
- `purchase_amount` - the amount of purchases (USD).

```python
browsing_time = [
    15.5, 16.2, 14.8, 15.3, 16.7, 27.0, 16.5, 15.8, 14.9, 16.1, 25.2, 24.8, 26.0,
    25.5, 16.2, 24.9, 25.7, 25.4, 24.6, 35.8, 40.3, 41.2, 39.5, 40.6, 39.8, 40.2,
    41.0, 40.7, 39.9, 20.1, 50.0, 80.0, 90.0
]

purchase_amount = [
    50.1, 48.5, 51.0, 49.3, 52.7, 90.9, 53.1, 49.5, 50.2, 48.8, 150.3, 152.0,
    151.2, 249.8, 150.6, 151.0, 150.5, 152.3, 149.9, 151.5, 100.1, 202.3, 199.5,
    201.6, 200.8, 199.7, 201.2, 202.0, 200.5, 301.1, 10.0, 500.0, 600.0
]
```

```python
import pandas as pd

data = pd.DataFrame({
    "browsing_time": browsing_time,
    "purchase_amount": purchase_amount,
    })
```

#2. ‚öôÔ∏è	Applying DBSCAN Clustering

```python
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

1st parameters combination:  

`eps` = 15  
`min_samples` = 5

```python
# apply DBSCAN clustering
dbscan = DBSCAN(eps = 15, min_samples = 5)
labels = dbscan.fit_predict(data)
```

```python
# plotting the result
plt.figure()
unique_labels = set(labels)

for label in unique_labels:
  if label == -1:   # noise points
    plt.scatter(data[labels == label]["browsing_time"], data[labels == label]["purchase_amount"], color = 'red', label = 'Noise', alpha = 0.6)
  else:
    plt.scatter(data[labels == label]["browsing_time"], data[labels == label]["purchase_amount"], label = f"Cluster {label + 1}", alpha = 0.7)


plt.title("DBSCAN Clustering")
plt.xlabel("Browsing Time")
plt.ylabel("Purchase Amount")
plt.legend()
plt.grid(True)
plt.show()
```

2nd parameters combination:  
`eps` = 10  
`min_samples` = 5

```python
# apply DBSCAN clustering
dbscan1 = DBSCAN(eps = 10, min_samples = 5)
labels = dbscan1.fit_predict(data)
```

```python
# plotting the result
plt.figure()
unique_labels = set(labels)

for label in unique_labels:
  if label == -1:   # noise points
    plt.scatter(data[labels == label]["browsing_time"], data[labels == label]["purchase_amount"], color = 'red', label = 'Noise', alpha = 0.6)
  else:
    plt.scatter(data[labels == label]["browsing_time"], data[labels == label]["purchase_amount"], label = f"Cluster {label + 1}", alpha = 0.7)


plt.title("DBSCAN Clustering")
plt.xlabel("Browsing Time")
plt.ylabel("Purchase Amount")
plt.legend()
plt.grid(True)
plt.show()
```

3rd parameters combination:  
`eps` = 8   
`min_samples` = 5

```python
# apply DBSCAN clustering
dbscan2 = DBSCAN(eps = 8, min_samples = 5)
labels = dbscan2.fit_predict(data)
```

```python
# plotting the result
plt.figure()
unique_labels = set(labels)

for label in unique_labels:
  if label == -1:   # noise points
    plt.scatter(data[labels == label]["browsing_time"], data[labels == label]["purchase_amount"], color = 'red', label = 'Noise', alpha = 0.6)
  else:
    plt.scatter(data[labels == label]["browsing_time"], data[labels == label]["purchase_amount"], label = f"Cluster {label + 1}", alpha = 0.7)


plt.title("DBSCAN Clustering")
plt.xlabel("Browsing Time")
plt.ylabel("Purchase Amount")
plt.legend()
plt.grid(True)
plt.show()
```

4th parameters combination:  
`eps` = 5   
`min_samples` = 5

```python
# apply DBSCAN clustering
dbscan3 = DBSCAN(eps = 5, min_samples = 5)
labels = dbscan3.fit_predict(data)
```

```python
# plotting the result
plt.figure()
unique_labels = set(labels)

for label in unique_labels:
  if label == -1:   # noise points
    plt.scatter(data[labels == label]["browsing_time"], data[labels == label]["purchase_amount"], color = 'red', label = 'Noise', alpha = 0.6)
  else:
    plt.scatter(data[labels == label]["browsing_time"], data[labels == label]["purchase_amount"], label = f"Cluster {label + 1}", alpha = 0.7)


plt.title("DBSCAN Clustering")
plt.xlabel("Browsing Time")
plt.ylabel("Purchase Amount")
plt.legend()
plt.grid(True)
plt.show()
```
